```python
import numpy as np
import matplotlib.pyplot as plt

from scipy.special import erfinv
from pydiffmap import diffusion_map as dm


%matplotlib inline
```

```python
# Parameters
n_evecs = 4
epsilon = 0.1
N = 50
k = 50
```

```python
# make points with equal density in gaussian space.
grid = np.linspace(0.,1.,N+2)[1:-1]
x = np.sqrt(2.)*erfinv(2.*grid-1.).reshape(-1,1)
grid_2 = np.linspace(0.,1.,2*N+2)[1:-1]
x2 = np.sqrt(2.)*erfinv(2.*grid_2-1.).reshape(-1,1)
```

```python
reload(dm)
mydmap = dm.DiffusionMap(n_evecs=n_evecs, epsilon=epsilon,
                         alpha=-.5, k=k)
mydmap.fit(x)
evecs = mydmap.transform(x2)
```

```python
for i in xrange(n_evecs):
    plt.plot(x2,evecs[:,i])
```
